```python
#напишем функцию для выделения 3 групп по удаленности от центра: до 5 км, до 15, свыше 15
def cityCenters_nearest_group (cityCenters_nearest):
    if cityCenters_nearest<=5000:
        return 'центр'
    if cityCenters_nearest<=15000:
        return 'спальные районы'
    return 'за КАДом'
#добавляем новый столбец, заполним названиями групп:
data['cityCenters_nearest_group'] = data['cityCenters_nearest'].apply(cityCenters_nearest_group)
#проверим, что получилось, выведем первые 5 строк таблицы: 
display(data.head())
```

```python
#переведем значения цены квартиры в  млн руб. с 3 знаками после зпт.:
data['last_price'] = (data['last_price']/1000000).round(3)
#добавим столбец с ценой за квадратный метр, в тыс. руб.: 
data['price_per_square_meter'] = (data['last_price'] / data['total_area'] * 1000).astype('int')
#выведем на экран первые 5 строк таблицы для проверки:
data.head()
```

```python
#Посмотрим значения квартилей цены за кв.м. по ранее выделенным группам по удаленности от центра
display('Центр:')
display(data.query('cityCenters_nearest_group == "центр"')['price_per_square_meter'].describe())
display('Cпальные районы:')
display(data.query('cityCenters_nearest_group == "спальные районы"')['price_per_square_meter'].describe())
display('За КАДом:')
data.query('cityCenters_nearest_group == "за КАДом"')['price_per_square_meter'].describe()
```

Значения цены за квадрат свыше Q3 будем считать "характерными" для квартир бизнес класса и выше, по этим квартирам, соответствено, будем отдельно считать высоту потолков. Остальные квартиры объединим в группу комфорт-класса и ниже. 

```python
#выделим группы для расчета и заполнения медианных значений по высоте потолков:
def cityCenters_and_priceSquare_group(row):
    price_per_square_meter = row['price_per_square_meter']
    cityCenters_nearest_group = row['cityCenters_nearest_group']
    if cityCenters_nearest_group == 'центр':
        if price_per_square_meter >= 153:
            return 'центр: бизнес-класс'
        return 'центр: комфорт-класс'
    if cityCenters_nearest_group == 'спальные районы':
        if price_per_square_meter >= 125:
            return 'спальные районы: бизнес-класс'
        return 'спальные районы: комфорт-класс'
    if cityCenters_nearest_group == 'за КАДом':
        if price_per_square_meter >= 96:
            return 'за КАДом: бизнес-класс'
        return 'за КАДом: комфорт-класс'
#добавляем новый столбец, заполним названиями групп:
data['cityCenters_and_priceSquare_group'] = data.apply(cityCenters_and_priceSquare_group, axis=1)
#выведем на экран первые 5 строк таблицы для проверки:
data.head()
```

```python
#посмотрим, какие средние значения высоты потолков у нас получились после разделения на группы:
data.groupby(by='cityCenters_and_priceSquare_group')['ceiling_height'].mean()
```

Существенная разница между комфорт и бизнес-классом заметна только в "спальных" районах. Связать это можно с тем, что в центре много старого фонда, и там остались высокие потолки. Новые проекты в центре- это в основном жилье бизнес - класса со стандартами бизнеса. 
За КАДом такие категории жилья как бизнес-класс неуместны по причине удаленности от центра, соответственно, и строят по стандартам комфорта (птолки 2,7 м). Ну а в спальных районах представлен и бизнес класс с 3-метровыми потолками, и комфорт. 

```python
#заменим пропуски значений в столбце ceiling_height средними значениями в соответствии с группами по столбцу cityCenters_and_priceSquare_group:
mean_ceiling_height = data.groupby(by='cityCenters_and_priceSquare_group')['ceiling_height'].transform('mean')
data['ceiling_height'] = data['ceiling_height'].fillna(mean_ceiling_height)
#проверим, остались ли пропуски:
data['ceiling_height'].isna().sum()
```

Но цена за квадратный метр будет выше в студии. Что и подтверждают найденные показатели коэффициента корреляции.
